```
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
```

```
m2_m1 = [1.,1.5,2.,2.5,3.,3.5,4.]
```

```
samples = 6
v1_v2 = np.zeros((len(m2_m1), samples))
for i in range(samples):
  v1_v2[:,i] = np.array([np.random.normal(-.5*m2_m1[f]+.5,.1) for f in range(len(m2_m1))])
```

```
stds = [np.std(v1_v2[i,:]) for i in range(len(m2_m1))]

for m, v in zip(m2_m1, v1_v2):
    plt.scatter([m] * len(v), v)

plt.text(2.5,0.25,'Mass Ratio vs Inverse Velocity Ratio (scatter)', horizontalalignment='center', fontsize=14)
plt.xlabel('$m_2/m_1$', fontsize=14)
plt.ylabel('$v_{1f}/v_{2f}$', fontsize=14)
plt.axis('scaled')
plt.xlim(.9,4.1)
plt.ylim(-1.6,0.2)
plt.savefig('scatter.png', dpi=300)
plt.show()
files.download('scatter.png') 
```

```
for i in range(len(v1_v2)):
  plt.errorbar(m2_m1[i], np.average(v1_v2[i,:]),yerr=stds[i], fmt='o', capsize=3)

plt.text(2.5,0.25,'Mass Ratio vs Inverse Velocity Ratio (bars)', horizontalalignment='center', fontsize=14)
plt.xlabel('$m_2/m_1$', fontsize=14)
plt.ylabel('$v_{1f}/v_{2f}$', fontsize=14)
plt.axis('scaled')
plt.xlim(.9,4.1)
plt.ylim(-1.6,0.2)
plt.savefig('bars.png', dpi=300)
plt.show()
files.download('bars.png') 
```
